#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO


ds_fil_precio_metros_sur['precio_normalizado'] = (ds_fil_precio_metros_sur['price'] - ds_fil_precio_metros_sur['price'].min()) / (ds_fil_precio_metros_sur['price'].max() - ds_fil_precio_metros_sur['price'].min())

poblaciones_sur1 = ds_sur['level5'].unique()
for i, poblacion in enumerate(poblaciones_sur1):
    nombres = ds_fil_precio_metros_sur[ds_fil_precio_metros_sur['level5'] == poblacion]
    plt.hist(nombres['precio_normalizado'], bins=20, alpha=0.5, label=poblacion)
plt.title('Histograma Normalizado de Precios por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

La normalización de los precios muestra que Alcorcón y Getafe tienen una mayor dispersión de precios, mientras que Fuenlabrada y Leganés tienen precios más homogéneos. Esto facilita la comparación entre poblaciones al eliminar las diferencias de escala.

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

```python
from ipyleaflet import Map, Marker, basemaps, AwesomeIcon

map = Map(center = (40.4168, -3.7038), zoom = 10, min_zoom = 1, max_zoom = 20, basemap=basemaps.OpenTopoMap)
```

```python
## HERE: plot the coordinates of the estates
colores = {
    'Leganés': 'green',  
    'Getafe': 'orange',   
    'Fuenlabrada': 'blue',
    'Alcorcón': 'red'
}
ds_fil_precio_metros_sur['color'] = ds_fil_precio_metros_sur['level5'].map(colores)
coordenadas_dict = ds_fil_precio_metros_sur[['level5', 'latitude', 'longitude', 'color']].to_dict(orient='records')
## PUT HERE YOUR CODE:
for poblacion in coordenadas_dict: 
        lat = poblacion['latitude'].replace(',', '.')
        lon = poblacion['longitude'].replace(',', '.') 
        color = poblacion['color']
        icon = AwesomeIcon(name='cloud', marker_color=color, icon_color='white', icon_size=[20, 20])
        marker = Marker(location=(lat, lon), icon=icon)
        map.add_layer(marker)
map

```
